Para estabelecer uma coluna como um índice basta usar set_index
- df.set_index('nome_da_coluna') # para uma ocluna
- df.set_index(['nome_da_coluna1', 'nome_da_coluna2', 'nome_da_coluna3']) # para várias colunas
- df.set_index([pd.Index][1,2,3,4,...], 'nome do indice') # para criar um índice novo do zero
