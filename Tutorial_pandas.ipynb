```python
#Formatação condicional, muda a cor dos valores abaixo de 91 nas colunas Overall e Potential
def low_values_red(value):
    color = 'red' if value < 92 else 'blak'
    return 'color: %s '%color

fifa[['Overall', 'Potential']].head(10).style.applymap(low_values_red)
```

```python
#Formatação condicional, mostra todas as colunas mas
# muda a cor dos valores abaixo de 91 apenas nas colunas Overall e Potential
fifa.head(10).style.applymap(low_values_red, subset=['Overall', 'Potential'])
```

```python
def highlight_max(s):
    is_max = s == s.max() # checa esse booleano (s == is_max) depois guarda em is_max
    return ['background-color: yellow' if v else '' for v in is_max]

fifa.head(10).style.apply(highlight_max, subset=['Age', 'Overall', 'Potential', 'Value'])
```

```python
# Utiliza as duas formatações
fifa[['Overall', 'Potential', 'Value']].head(10).style.\
    applymap(low_values_red).\
    apply(highlight_max)
```

## Convertendo str para float

```python
# retirando linhas null para 0 lb
fifa['Weight'] = fifa['Weight'].fillna('0lbs')
```

```python
def ajuste_peso(weigth):
    peso = weigth.replace('lbs', '')
    if (len(peso) > 10):
        return float(peso[:10])
    else:
        return float(peso)
```

```python
teste = '120lbs'
ajuste_peso(teste)
```

```python
fifa['Weigth'] = fifa['Weight'].apply(ajuste_peso)
```

### Inserir parametros

```python
# O asterisco serve para que a função fique aberta em relação ao númeor de parametros
def extra_parametros(var1, *tupla): 
    print("O parametro declarado é: " + str(var1))
    print("os parametros extras são: ")
    for var in tupla:
        print(str(var))
    return;
```

```python
extra_parametros(5,3,6,81,3)
```
